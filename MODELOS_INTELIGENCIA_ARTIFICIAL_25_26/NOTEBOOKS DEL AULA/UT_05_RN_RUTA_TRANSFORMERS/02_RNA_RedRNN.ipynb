## RED NEURONAL RECURRENTE - (*RNN - Recurrent Neural Network o Red de ELMAN*)   

Las redes de Elman son el modelo mГЎs simple de Red Neuronal Recurrente (RNN). Tienen la misma estructura que las redes neuronales vistas hasta ahora, salvo por una Гєnica circunstancia: se permite que cada neurona se **retroalimente a sГ­ misma**.

<img src="./img/Elman.jpg">   

Donde $h_t$ es el estado de la neurona en el momento $t$, $h_{t-1}$ su estado en el momento inmediatamente anterior; $w_i$ representa los pesos sinГЎpticos, $d_i$ los valores de activaciГіn de las neuronas de la capa anterior y $b$ el sesgo(NO APARECE EN LA IMAGEN). Como se puede observar, existe un tГ©rmino extra $Uh_{t-1}$ que no existe en el caso de las redes no recurrentes, y que en este caso permite que cada neurona se excite a sГ­ misma.    

Como se puede ver, por tanto, en las neuronas del modelo de Elman se forma un pequeГ±o bucle de retroalimentaciГіn, mediante el cual el axГіn de la neurona excita una de sus propias dendritas. El peso sinГЎptico de esta conexiГіn de retroalimentaciГіn es $U$.

El valor de activaciГіn en el momento anterior $h_{t-1}$ contribuye a la suma total de excitaciones de la cГ©lula $(x)$, a travГ©s del coeficiente $U$. Este coeficiente funcionarГЎ de forma equivalente a un peso sinГЎptico.

## EJEMPLO DE RNN


Este es un ejemplo sencillo de una RNN que va a permitir llevar a cabo la predicciГіn de caracteres usando un conjunto de datos (cadena de texto).

### IMPORTAR LIBRERГЌAS NECESARIAS

```python
import warnings
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.optimizers import Adam
```

Desactivamos los WARNINGS

```python
warnings.filterwarnings("ignore")
```

### DEFINICIГ“N DEL CONJUNTO DE DATOS Y CREACIГ“N DEL VOCABULARIO.

Generamos el conjunto de datos a partir de una frase sencilla y, seguidamente, creamos un vocabulario ordenando el conjunto de carГЎcteres que componen la frase.   


```python
# Definimos el conjunto de datos (secuencia de texto)
text = "hola mundo"
# Crear un vocabulario de caracteres
vocab = sorted(set(text))
vocab_size = len(vocab)

print("Vocabulario: ", vocab)
print("TamaГ±o del vocabulario: ", vocab_size)
```


Seguidamente se crea un diccionario mapeando los caracteres del vocabulario con Г­ndices para, a continuaciГіn, transformar la secuencia de caracteres que conforman la frase en una secuencia de indices.

```python
# Crear un diccionario de mapeo de caracteres a Г­ndices
char_to_idx = {char: idx for idx, char in enumerate(vocab)}
idx_to_char = np.array(vocab)
print("Mapeo de caracteres a Г­ndices: ", char_to_idx)
print("Mapeo de Г­ndices a caracteres: ", idx_to_char)

# Convertir la secuencia de texto a una secuencia de Г­ndices
text_as_int = np.array([char_to_idx[c] for c in text])
print("Texto como Г­ndices: ", text_as_int)
```

### PREPARACIГ“N DE DATOS

Para preparar los datos, en el siguiente bloque de cГіdigo se establece la longitud de la secuencia de entrada (tambiГ©n denominada "ventana") a un valor de 4 caracteres, por ejemplo: "mund" -> "o".      

En este caso, para el entrenamiento se generan un total de 6 ejemplos para la secuencia de partida: "hola mundo".    

Finalizado este proceso, se dispondrГЎ de los datos de entrenamiento formando un conjunto de pares de secuencia de entrada y salida.

```python
# Preparar los datos de entrenamiento y etiquetas
seq_length = 4 # TamaГ±o de la ventana = 4. Se toman 4 caracteres como entrada y se predice el siguiente caracter
examples_per_epoch = len(text) - seq_length 
```

```python
# Crear las secuencias de entrada y salida
inputs = np.array([text_as_int[i:i+seq_length] for i in range(examples_per_epoch)])
targets = np.array([text_as_int[i+seq_length] for i in range(examples_per_epoch)])

# Reshape para cumplir con el formato esperado por la RNN
inputs = np.reshape(inputs, (examples_per_epoch, seq_length, 1)) # se corresponde con (batch_size, seq_length, input_dim)

# Usar tf.data.Dataset para manejar los datos
# y crear un dataset a partir de los tensores de entrada y salida (inputs, targets)
dataset = tf.data.Dataset.from_tensor_slices((inputs, targets)) #from_tensor_slices crea un dataset a partir de tensores, donde cada tensor es una muestra de entrada y salida
# batch(1) indica que se toma un solo ejemplo por batch y drop_remainder=True indica que se descartan los ejemplos que no se ajustan al tamaГ±o del batch
dataset = dataset.batch(1, drop_remainder=True) 
```

```python
print("Ejemplos por Г©poca: ", examples_per_epoch)
print("Entradas: ", inputs.shape)
print("Etiquetas: ", targets.shape)
```

### DISEГ‘O DE LA ARQUITECTURA DE LA RNN

```python
# Crear el modelo RNN
model = Sequential([
    SimpleRNN(50,                           # NГєmero de unidades en la capa oculta (50)
              input_shape=(seq_length, 1),  # TamaГ±o de la ventana y nГєmero de caracterГ­sticas (1)
              return_sequences=False),      #return_sequences=False indica que solo se devuelve la salida de la Гєltima capa
    
    Dense(vocab_size,                       # NГєmero de unidades en la capa de salida (vocab_size, tamaГ±o del vocabulario)
          activation='softmax')             # FunciГіn de activaciГіn de la capa de salida (softmax), ya que se trata de un problema de clasificaciГіn multiclase.
])

```

### COMPILACIГ“N

```python
# Compilar el modelo
model.compile(optimizer=Adam(),                         # Optimizador Adam por defecto
              loss='sparse_categorical_crossentropy')   #sparse_categorical_crossentropy ya que se trata de un problema de clasificaciГіn multiclase
```

### ENTRENAMIENTO

```python
# Entrenar el modelo
model.fit(dataset, epochs=100)
```

### PREDICCIГ“N

```python
# FunciГіn para predecir el siguiente carГЎcter en una secuencia dada
def predict_next_char(model, input_text):
    input_eval = np.array([char_to_idx[c] for c in input_text])         # Convertir la entrada a Г­ndices
    input_eval = np.reshape(input_eval, (1, len(input_eval), 1))        # Reshape para cumplir con el formato esperado por la RNN
    prediction = model.predict(input_eval)                              # Realizar la predicciГіn
    predicted_idx = np.argmax(prediction)                               # Obtener el Г­ndice del carГЎcter predicho
    return idx_to_char[predicted_idx]                                   # Obtener el carГЎcter correspondiente al Г­ndice predicho
```

```python
# Probar el modelo
input_text = "mund"
predicted_char = predict_next_char(model, input_text)
print(f"Entrada: '{input_text}' -> Siguiente carГЎcter predicho: '{predicted_char}'")
```

## **NOTA SOBRE LAS RNN Y LAS TAREAS SOBRE SERIES TEMPORALES**   

Las Redes Neuronales Recurrentes (RNN) se utilizan ampliamente para tareas con series temporales, aunque ciertas variantes son mГЎs adecuadas que las RNNs bГЎsicas.

**ВїPor quГ© las RNNs para series temporales?**

* **Dependencia temporal:**
    * Las series temporales son secuencias de datos donde el orden es crucial. Las RNNs estГЎn diseГ±adas para manejar datos secuenciales, ya que mantienen una "memoria" de las entradas anteriores. Esto les permite capturar dependencias temporales, es decir, cГіmo los valores pasados influyen en los valores futuros.
    * Por ejemplo, para predecir el precio de una acciГіn, es importante considerar los precios anteriores, ya que suelen existir patrones y tendencias a lo largo del tiempo.
* **Procesamiento de secuencias de longitud variable:**
    * Las RNNs pueden procesar secuencias de longitud variable, lo que es Гєtil para series temporales que pueden tener diferentes duraciones.

**Tipos de RNNs mГЎs utilizados para series temporales:**

* **Redes LSTM (Long Short-Term Memory):**
    * Las LSTM son una variante de las RNNs que abordan el problema del "desvanecimiento del gradiente", que dificulta el aprendizaje de dependencias a largo plazo.
    * Las LSTM utilizan "compuertas" que controlan el flujo de informaciГіn, lo que les permite recordar informaciГіn relevante durante perГ­odos prolongados.
    * Son muy efectivas para series temporales con patrones complejos y dependencias a largo plazo, como la predicciГіn del precio de acciones o el anГЎlisis del lenguaje natural.
* **Redes GRU (Gated Recurrent Units):**
    * Las GRU son una versiГіn simplificada de las LSTM que tambiГ©n abordan el problema del desvanecimiento del gradiente.
    * Tienen menos compuertas que las LSTM, lo que las hace computacionalmente mГЎs eficientes.
    * Las GRU suelen ofrecer un rendimiento similar a las LSTM en muchas tareas de series temporales.

**CONCLUSIГ“N:**

* Las RNNs, especialmente las LSTM y GRU, son herramientas poderosas para el anГЎlisis y la predicciГіn de series temporales debido a su capacidad para capturar dependencias temporales.
* Si bien las RNN basica se pueden usar, LSTM y GRU dan mejor rendimiento con series temporales.
