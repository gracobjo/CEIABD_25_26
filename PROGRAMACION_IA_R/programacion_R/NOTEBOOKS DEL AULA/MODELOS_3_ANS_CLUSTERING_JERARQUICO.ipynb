### Análisis de segmentación de clientes utilizando Clustering Jerárquico   

En este ejemplo, utilizaremos el algoritmo de Clustering Jerárquico para segmentar una base de datos de clientes en función de sus características. El objetivo es identificar grupos de clientes similares para personalizar estrategias de marketing.

#### Carga de Datos y Preprocesamiento

```r
library(dplyr)
library(ggplot2)

data <- read.csv("/home/jordi/Documentos/Ribera/Curso_25_26/CEIABD/CEIABD_25_26/PROGRAMACION_IA_R/programacion_R/NOTEBOOKS DEL AULA/datasets/Mall_Customers.csv")

sum(is.na(data))
```

#### Escalado de Datos   

Se procede a escalar los datos para asegurar que todas las características contribuyan de manera equitativa al cálculo de distancias.

```r
# Renombramos las columnas para mayor claridad y consistencia
colnames(data) <- c("IDCliente", "Genero", "Edad", "IngresosAnuales", "PuntuacionGasto")

# Eliminamos la columna CustomerID ya que no es relevante para el clustering
data <- data[, c("Edad", "IngresosAnuales", "PuntuacionGasto")]

# El escalado se aplica solo a las columnas numéricas (Age, AnnualIncome, SpendingScore)
data_scaled <- scale(data)

# Visualizamos las primeras filas del conjunto de datos escalados
head(data_scaled)
```
