### Fase 2: Simulando un ataque (Evidenciando la Inmutabilidad)
**Instrucción para el alumno:** En esta fase se actuará como atacantes. ¿Qué pasa si intentamos cambiar la historia del primer bloque de forma maliciosa?

Añadir este código a continuación del anterior y ejecutar:

```python
# Un atacante altera el historial del Bloque A
block_A.history = 'Alice hates cats'

# Recalculamos cómo se vería el hash del Bloque B ahora, dado el cambio en A
nuevo_hash_padre_para_B = hashlib.sha256(json.dumps(block_A.__dict__).encode('utf-8')).hexdigest()

# Recalculamos cómo se vería el hash del Bloque C si actualizamos B
block_B.parent_hash = nuevo_hash_padre_para_B
nuevo_hash_padre_para_C = hashlib.sha256(json.dumps(block_B.__dict__).encode('utf-8')).hexdigest()

print("\n--- DESPUÉS DEL ATAQUE ---")
print("Nuevo Hash generado para el Bloque C:")
print(nuevo_hash_padre_para_C)
```

**Análisis en clase:** Al ejecutarlo, se verá que los hashes resultantes son completamente diferentes. 
Comentar cómo, en una red real de miles de bloques, un atacante tendría que recalcular el hash de *todos* los bloques subsecuentes para evitar ser detectado. Esto evidencia la inmutabilidad.

### Fase 3: Introduciendo la Minería (Prueba de Trabajo)
**Instrucción para el alumno:** Recalcular hashes rápidamente es fácil para una computadora moderna.     

Para evitar que el atacante del paso anterior recalcule toda la cadena en milisegundos, las redes como Bitcoin exigen una **Prueba de Trabajo**. Vamos a obligar al sistema a que el hash válido deba empezar obligatoriamente con cinco ceros (`00000`).

Crear un nuevo archivo y ejecutar el siguiente script de fuerza bruta:

```python

import hashlib
import time

print("Iniciando la minería del nuevo bloque...")
start_time = time.time()

# Datos del bloque que queremos añadir a la red
payload = b'{"history": "David loves turtles", "parent_id": 3, "id": 4}'

# El minero prueba números (nonce) mediante fuerza bruta hasta resolver el acertijo
for i in range(10000000):
    nonce = str(i).encode('utf-8')
    # Se concatena la información del bloque con el nonce
    result = hashlib.sha256(payload + nonce).hexdigest()
    
    # Comprobamos si el hash cumple la condición impuesta (5 ceros)
    if result[0:5] == '00000':
        end_time = time.time()
        print("\n¡BLOQUE MINADO CON ÉXITO!")
        print(f"Nonce encontrado (solución): {i}")
        print(f"Hash resultante: {result}")
        print(f"Tiempo de minado: {round(end_time - start_time, 2)} segundos")
        break
```

**Análisis en clase:** 
*   Al ejecutar el script de minería, se observará que la computadora tarda un tiempo tangible en procesar miles de combinaciones hasta encontrar el "nonce" correcto que genera el patrón de cinco ceros. 
*   **Reto extra (Dificultad de red):** Se debe cambiar la condición de `00000` a `000000` (seis ceros) en la instrucción `if result[0:6] == '000000':` y observar cómo el tiempo necesario para minar el bloque aumenta exponencialmente.
