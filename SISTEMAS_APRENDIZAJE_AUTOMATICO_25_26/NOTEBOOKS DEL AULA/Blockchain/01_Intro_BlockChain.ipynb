# Introducción al Blockchain

La tecnología **blockchain (o cadena de bloques)** se define como una **base de datos de solo adición (append-only) conformada por bloques que están enlazados entre sí mediante funciones hash**. Una vez que un dato se introduce en esta base de datos, queda registrado permanentemente y no puede ser modificado ni alterado, ya que no existe un "botón de deshacer".

Para comprenderla detalladamente, es fundamental desglosar sus componentes y mecanismos principales:

**1. Estructura de Bloques y Enlaces Hash:**
La información se agrupa en bloques, donde cada bloque contiene una lista de transacciones (que pueden ser transferencias de valor o cambios de estado en un programa) aseguradas mediante criptografía. Además de las transacciones, un bloque almacena otros datos críticos: la ***marca de tiempo*** de cuando fue añadido, la ***dificultad de la red*** y, lo más importante, el **hash del bloque padre (el bloque anterior)**. Al incluir el hash del bloque anterior, se forma una cadena inquebrantable. Si alguien intentara alterar la información de un bloque antiguo, el hash de ese bloque cambiaría por completo, lo que invalidaría el hash de su hijo y de todos los bloques subsecuentes, haciendo que el fraude sea evidente y extremadamente difícil de ocultar.

**2. Criptografía Asimétrica:**
Para identificar y autenticar a los participantes, blockchain utiliza un par de claves: una clave privada (secreta) y una clave pública (compartida con todos). Esto **permite que los usuarios firmen criptográficamente sus datos e interacciones**, garantizando que solo el propietario de los fondos o el emisor autorizado de un mensaje pueda realizar una transacción sin que terceros no autorizados la alteren. 

**3. Descentralización y Consenso:**
A diferencia de los sistemas tradicionales, blockchain se distribuye entre multitud de participantes (conocidos como nodos, mineros o validadores) que mantienen una copia idéntica de la base de datos. Dado que la red es global y los mensajes tardan en propagarse, los nodos utilizan un **mecanismo de consenso** para sincronizarse y decidir cuál es el siguiente bloque legítimo que se debe añadir. La regla principal en este consenso suele ser aceptar y adoptar siempre **la cadena más larga**, ya que es la que representa el mayor esfuerzo computacional o participación de la mayoría.

**4. Prevención contra Fraudes y Mecanismos de Validación:**
Para añadir un nuevo bloque y asegurar la red, los validadores deben demostrar su compromiso, lo cual evita los ataques de doble gasto (double-spending) y la manipulación del historial.
*   **Prueba de Trabajo (Proof of Work):** Utilizada por redes como *Bitcoin*, obliga a los mineros a gastar recursos computacionales para resolver acertijos matemáticos complejos mediante fuerza bruta. Solo el primero en resolver el puzzle tiene derecho a proponer un bloque y recibir una recompensa.
*   **Prueba de Participación (Proof of Stake):** Utilizada por *Ethereum* en la actualidad. En lugar de gastar energía en cálculos, los validadores bloquean sus propias criptomonedas (como garantía o "stake") para proponer y verificar bloques. Si el sistema detecta que un validador intenta hacer trampa, se le confiscan los fondos, lo que garantiza económicamente su honestidad.

**5. Evolución del Almacenamiento:**
Mientras que originalmente (en el caso de *Bitcoin*) blockchain fue diseñada estrictamente para **prevenir trampas en el almacenamiento y envío de valor** monetario de forma descentralizada, la tecnología ha evolucionado enormemente con *Ethereum*. *Ethereum* convirtió blockchain en una **plataforma para computación segura y transparente**, permitiendo almacenar "contratos inteligentes" (código de programación) que no pueden ser censurados ni manipulados ni siquiera por sus propios desarrolladores.

### EJEMPLO SENCILLO      

La implementación fundamental de una blockchain se asemeja a una lista enlazada (linked list), en la cual cada bloque contiene información y guarda una firma digital (hash) del bloque que lo precede.

```python
# Imports de librerias necesarias

import hashlib
import json
```

```python
# Definición básica de la estructura del bloque
class Block:
    id = None
    history = None
    parent_id = None
    parent_hash = None
```

```python
# 1. Creación del primer bloque (Bloque Génesis)
block_A = Block()
block_A.id = 1
block_A.history = 'Nelson likes cat'
```

```python
# 2. Creación del segundo bloque enlazado al primero
block_B = Block()
block_B.id = 2
block_B.history = 'Marie likes dog'
block_B.parent_id = block_A.id
# Se calcula el hash del bloque padre (block_A) y se almacena en block_B
block_B.parent_hash = hashlib.sha256(json.dumps(block_A.__dict__).encode('utf-8')).hexdigest()
```

```python
# 3. Creación del tercer bloque enlazado al segundo
block_C = Block()
block_C.id = 3
block_C.history = 'Marie likes dog'
block_C.parent_id = block_B.id
# Se calcula el hash del bloque padre (block_B) y se almacena en block_C
block_C.parent_hash = hashlib.sha256(json.dumps(block_B.__dict__).encode('utf-8')).hexdigest()
```

Este primer fragmento demuestra cómo la cadena de bloques previene las trampas. Si alguien intentara alterar la historia del block_A de forma maliciosa, el hash resultante de ese bloque cambiaría por completo. Esto invalidaría inmediatamente el parent_hash almacenado en block_B, rompiendo la cadena y obligando al atacante a recalcular todos los hashes de los bloques subsecuentes para que el cambio pasara desapercibido.     

Sin embargo, el hashing por sí solo no es suficiente para evitar que alguien con una computadora moderna altere toda la cadena rápidamente. Para garantizar la seguridad del sistema y regular cómo se añaden nuevos bloques, se introduce la **Prueba de Trabajo (Proof of Work)**. En lugar de añadir un bloque directamente, se exige a los participantes (mineros) resolver un acertijo computacional: encontrar un *nonce* (un número arbitrario) que, al concatenarse con los datos del bloque y aplicarles la función hash, produzca un resultado que comience con una cantidad específica de ceros (por ejemplo, cinco ceros).     

A continuación, se ilustra el código en Python que simula este esfuerzo de minería por fuerza bruta para añadir un nuevo bloque candidato a la cadena:

```python
import hashlib

# Datos serializados del nuevo bloque candidato
payload = b'{"history": "Sky loves turtle", "parent_id": 3, "id": 4}'

# Fuerza bruta para encontrar el hash válido (que empiece con '00000')
for i in range(10000000):
    nonce = str(i).encode('utf-8')
    # Se concatena el payload con el nonce propuesto y se calcula el hash
    result = hashlib.sha256(payload + nonce).hexdigest()
    
    if result[0:5] == '00000':
        # ¡El minero logró resolver el acertijo!
        print("Nonce encontrado:", i)
        print("Hash resultante:", result)
        break
```

## 1. Arquitectura y Fundamentos de DLT   

***Blockchain*** se clasifica dentro de las Tecnologías de Libro Mayor Distribuido (DLT). A diferencia de una base de datos centralizada, donde un administrador controla la información, en una red P2P (peer-to-peer) cada nodo mantiene una copia idéntica del registro.   

La idea fundamental de esta tecnología es mantener un libro mayor (o libro de contabilidad) distribuido, el cual funciona como un registro de transacciones que es completamente descentralizado y a prueba de manipulaciones. Al ser un sistema descentralizado, todos los nodos que participan en la red poseen una copia idéntica del registro de transacciones, en lugar de depender de un único servidor central.

Sus características y ventajas principales incluyen:
- *Eliminación de intermediarios*: Un libro mayor distribuido hace posible realizar transacciones seguras y transparentes entre múltiples participantes de la red sin la necesidad de depender de intermediarios o terceros de confianza.
- *Inmutabilidad y seguridad*: Actúa como un registro a prueba de manipulaciones, lo que garantiza la integridad de la información y la protección contra fraudes una vez que una transacción ha sido validada.
- *Versatilidad de redes*: Las plataformas DLT han experimentado un rápido crecimiento y ofrecen varios tipos de redes según el nivel de acceso deseado: pueden ser públicas, privadas o permisionadas.
- *Integración de lógica de negocio*: En cada una de estas redes se pueden incorporar capacidades de contratos inteligentes (smart contracts) diseñados a la medida para resolver casos de uso específicos.

Por otra parte, a nivel de estructura, para implementar un sistema blockchain (por ejemplo, a nivel local) debemos tener en consideración los siguientes elementos:

- **Estructura del Bloque**: Cada bloque es una unidad de almacenamiento que contiene transacciones y metadatos críticos.   
- **Encadenamiento Criptográfico**: Los bloques se enlazan mediante el campo *previous_hash* (hash del bloque anterior). Si un atacante modifica un dato en el bloque $N$, el *hash* de dicho bloque cambia, invalidando el enlace con el bloque $N+1$ y todos los sucesivos.   
- **Bloque Génesis**: Es el primer bloque de la cadena (índice 0), que no tiene un *previous_hash* ya que no hay nada antes de él.    

Si vamos a considerar una estructura distribuida, ademas es necesario tener en cuenta también:       

- *Mecanismo de consenso*: cómo los nodos de la red se ponen de acuerdo sobre qué bloque es válido (Proof of Work, Proof of Stake, etc.). Sin esto se tiene una cadena local, no una blockchain distribuida.
- *Función hash criptográfica*: hay que especificar qué algoritmo se usa (SHA-256 es el canónico), ya que es el pegamento que hace funcionar el encadenamiento.
- *Validación de la cadena*: la lógica para verificar que ningún bloque ha sido manipulado, que es consecuencia directa de tu punto sobre el encadenamiento pero merece ser explícita.
- *Inmutabilidad práctica*: el coste computacional o de coordinación que hace inviable reescribir la historia (relacionado con el consenso).

## 2. Criptografía Aplicada    

La seguridad de la cadena descansa en dos pilares criptográficos:   
- **Funciones Hash (SHA-256/SHA-3)**: Son funciones unidireccionales que generan una "huella digital" de tamaño fijo para cualquier entrada. Son deterministas (misma entrada, mismo hash) y presentan un efecto avalancha (cambios mínimos en la entrada alteran radicalmente el hash).   
- **Criptografía Asimétrica y Firmas Digitales**: Los usuarios se identifican mediante una clave pública (similar a un número de cuenta) y firman transacciones con su clave privada. Esto permite verificar la autoría y la integridad de la transacción sin revelar la clave secreta.

## 3. Algoritmos de Consenso: El Corazón de la Red   

El consenso es el mecanismo que permite a nodos que no se conocen confiar en la validez de los datos.   

- **Proof-of-Work (PoW)**: Utilizado en Bitcoin, requiere que los mineros resuelvan un puzzle matemático encontrando un valor llamado ***nonce***. Este valor, al ser hasheado con el resto de la cabecera, debe producir un hash que cumpla con una dificultad específica (ej. empezar con $N$ ceros).    
- **Proof-of-Stake (PoS)**: Sustituye el gasto energético por una "apuesta" o depósito de monedas. Los validadores se eligen en función de su participación económica, lo que mejora drásticamente la escalabilidad y reduce el impacto ambiental.

## 4. Implementación Técnica en Python.   

A continuación, se muestra una simulación de la lógica de un bloque y su validación, usando una estructura más sólida y fiable:

### 1.- Importar librerías necesarias

```python
import hashlib
import json
from time import time
```

### 2.- Definición de la clase Blockchain (Bloques)

```python
class Blockchain:
    def __init__(self):
        self.chain = []
        self.pending_transactions = []
        # Crear el bloque génesis
        self.create_block(nonce=100, previous_hash='0')

    def create_block(self, nonce, previous_hash):
        """Crea un nuevo bloque y lo añade a la cadena"""
        block = {
            'index': len(self.chain) + 1,
            'timestamp': time(),
            'transactions': self.pending_transactions,
            'nonce': nonce,
            'previous_hash': previous_hash,
        }
        # Resetear transacciones pendientes
        self.pending_transactions = []
        self.chain.append(block)
        return block

    def add_transaction(self, sender, receiver, amount):
        """Añade una interacción/transacción a la lista de espera"""
        self.pending_transactions.append({
            'sender': sender,
            'receiver': receiver,
            'amount': amount,
        })
        return self.last_block['index'] + 1

    @staticmethod
    def hash(block):
        """Crea un hash SHA-256 de un bloque"""
        block_string = json.dumps(block, sort_keys=True).encode()
        return hashlib.sha256(block_string).hexdigest()

    @property
    def last_block(self):
        return self.chain[-1]

    def proof_of_work(self, last_nonce):
        """
        Algoritmo de Consenso simple:
        Buscar un número 'p' tal que hash(pp') contenga 4 ceros a la izquierda
        """
        nonce = 0
        while self.valid_proof(last_nonce, nonce) is False:
            nonce += 1
        return nonce

    @staticmethod
    def valid_proof(last_nonce, nonce):
        guess = f'{last_nonce}{nonce}'.encode()
        guess_hash = hashlib.sha256(guess).hexdigest()
        return guess_hash[:4] == "0000"

```

### 3.- Ejecución de la Interacción.    

#### 3.1 - Instanciar la red

```python

my_blockchain = Blockchain()
```

#### 3.2 - Simulación de interacciones

```python
print("Añadiendo transacciones...")
my_blockchain.add_transaction("Alice", "Bob", 50)
my_blockchain.add_transaction("Data_Sensor_01", "AI_Model_Trainer", 0.002)
```

#### 3.3 - Proceso de minería - (CONSENSO)

```python

print("Iniciando minería...")
last_block = my_blockchain.last_block
last_nonce = last_block['nonce']
nonce = my_blockchain.proof_of_work(last_nonce)

```

#### 3.4 - Forjado del nuevo bloque

```python
previous_hash = my_blockchain.hash(last_block)
block = my_blockchain.create_block(nonce, previous_hash)

print(f"\nNuevo bloque minado con éxito:")
print(json.dumps(block, indent=2))
print(f"\nCadena completa: {json.dumps(my_blockchain.chain, indent=2)}")
```

## Análisis Técnico.    

- **Inmutabilidad y Verificación**: En el método `hash()`, utilizamos `sort_keys=True`. En Big Data, el orden de los factores sí altera el producto criptográfico. Si cambiamos un solo bit en una transacción de un bloque antiguo, el hash resultante sería totalmente distinto, rompiendo la cadena.

- **Costo Computacional (Dificultad)**: En el método `valid_proof`, la condición `guess_hash[:4] == "0000"` determina la dificultad. En redes reales como Bitcoin, esta dificultad se ajusta dinámicamente para mantener un tiempo de bloque constante, enfrentando el problema de la escalabilidad.

- **Transacciones Pendientes (Mempool)**: Las transacciones no se graban instantáneamente. Pasan por un estado de "pendientes" hasta que el algoritmo de consenso las convalida. Esto introduce una latencia que debéis considerar al diseñar arquitecturas de IA en tiempo real que consuman datos de una blockchain.
